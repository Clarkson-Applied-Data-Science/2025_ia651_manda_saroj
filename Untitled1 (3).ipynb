```python
# Ensure 'Date' column is in datetime format
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'], errors='coerce')

# Remove non-numeric characters from 'Price', 'Vol.', and 'Change %' and convert them to numeric
bitcoin_df['Price'] = bitcoin_df['Price'].replace({'\$': '', ',': '', 'K': 'e3', 'M': 'e6', 'B': 'e9'}, regex=True)
bitcoin_df['Vol.'] = bitcoin_df['Vol.'].replace({'\$': '', ',': '', 'K': 'e3', 'M': 'e6', 'B': 'e9'}, regex=True)
bitcoin_df['Change %'] = bitcoin_df['Change %'].replace({'%': '', ',': ''}, regex=True)

# Convert to numeric
bitcoin_df['Price'] = pd.to_numeric(bitcoin_df['Price'], errors='coerce')
bitcoin_df['Vol.'] = pd.to_numeric(bitcoin_df['Vol.'], errors='coerce')
bitcoin_df['Change %'] = pd.to_numeric(bitcoin_df['Change %'], errors='coerce')

# Check for any NaN values after cleaning
print(bitcoin_df.isnull().sum())
```

```python
# Create Lag Features
bitcoin_df['Lag_1'] = bitcoin_df['Price'].shift(1)
bitcoin_df['Lag_2'] = bitcoin_df['Price'].shift(2)

# Drop any rows with NaN values created by lagging
bitcoin_df.dropna(inplace=True)
```

```python
train_size = int(len(bitcoin_df) * 0.8)
train, test = bitcoin_df['Price'][:train_size], bitcoin_df['Price'][train_size:]
X_train, X_test = bitcoin_df[['Lag_1', 'Lag_2']][:train_size], bitcoin_df[['Lag_1', 'Lag_2']][train_size:]
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
model = ARIMA(train, exog=X_train, order=(1, 0, 0))  # ARIMA(1, 0, 0) with exogenous variables (lag features)
model_fit = model.fit()

# Output model summary
print(model_fit.summary())
```

```python
# Make predictions on the test data
predictions = model_fit.forecast(steps=len(test), exog=X_test)

# Plot the predictions vs actual values
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, predictions, label='Predicted', color='red')
plt.title('ARIMA Model Predictions vs Actual')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test, predictions))
print(f'RMSE for ARIMA model: {rmse}')
```

```python
from pmdarima import auto_arima

# Fit an AutoARIMA model to select the best hyperparameters
auto_model = auto_arima(train, exogenous=X_train, seasonal=False, trace=True)

# Output the best model summary
print(auto_model.summary())
```

```python
# Residuals plot
residuals = model_fit.resid

plt.figure(figsize=(12, 6))
plt.subplot(211)
plt.plot(residuals)
plt.title('Residuals from ARIMA Model')

plt.subplot(212)
plt.hist(residuals, bins=20)
plt.title('Histogram of Residuals')

plt.show()
```

```python
# Define the forecast steps
forecast_steps = 10  # for example, forecast for the next 10 days

# Ensure you have the last known lag values from the test set for the initial forecast
latest_lagged_values = bitcoin_df[['Lag_1', 'Lag_2']].iloc[-1:].values  # Use the last row of lag features
latest_lagged_values = np.tile(latest_lagged_values, (forecast_steps, 1))  # Replicate this row for the forecast steps

# Initialize a list to store the predictions
predictions = []

# Loop to update lag values and forecast each future step
for i in range(forecast_steps):
    # Update lag values based on previous predictions (if applicable)
    new_lag_1 = predictions[i - 1] if i > 0 else latest_lagged_values[i, 0]  # Update lag 1
    new_lag_2 = predictions[i - 2] if i > 1 else latest_lagged_values[i, 1]  # Update lag 2
    next_lagged_values = np.array([[new_lag_1, new_lag_2]])  # Create new lag values for this step
    
    # Make the forecast for this step with updated lag values
    next_pred = model_fit.forecast(steps=1, exog=next_lagged_values)
    
    # Append the prediction to the list
    predictions.append(next_pred[0])

# Create a date range for the forecast period (based on the last date of the test data)
last_date = test.index[-1]
forecast_dates = pd.date_range(start=last_date, periods=forecast_steps + 1, freq='D')[1:]

# Plot the actual Bitcoin prices, predicted Bitcoin prices, and forecasted Bitcoin prices
plt.figure(figsize=(12, 6))
plt.plot(test.index, test, label='Actual Bitcoin Prices', color='blue')  # Actual data
plt.plot(test.index, predictions, label='Predicted Bitcoin Prices', color='red')  # Predicted data
plt.plot(forecast_dates, predictions, label=f'Forecasted Bitcoin Prices ({forecast_steps} days)', color='green')  # Forecasted data

# Customize the plot
plt.title(f'ARIMA Model: Actual, Predicted, and Forecasted Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Bitcoin Price')
plt.legend()
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()

# Output the predicted Bitcoin prices for the future (after 21st April)
print(f"Predicted Bitcoin Prices for the next {forecast_steps} days: {predictions}")
```
